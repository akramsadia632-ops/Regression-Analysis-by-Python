```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#R&D Spend vs Profit
```

```python
plt.scatter(df["R&D Spend"], df["Profit"], color='red', marker='o', alpha=0.7)
plt.xlabel("R&D Spend")
plt.ylabel("Profit")
plt.title("R&D Spend vs Profit")
plt.show()
```

```python
# put input variables in X
# and output which are independent variables in Y
```

```python
X = df[["R&D Spend"]]
Y = df["Profit"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
Lr = LinearRegression()
Lr.fit(X_train, Y_train)
```

```python
Y_pred=Lr.predict(X_test)
```

```python
# finding evaluation matrix
```

```python
r2_score = r2_score(Y_test, Y_pred)
MSE = mean_squared_error(Y_test, Y_pred)
```

```python
import gradio as gr
import joblib
import numpy as np

# Load your trained model
model = joblib.load(r"C:\\Users\\ADMIN\\OneDrive\\Documents\\Machine Learning\\model.pkl")

# Define the prediction function
def predict_profit(rd_spend):
    features = np.array([[rd_spend]])
    prediction = model.predict(features)
    return f"Predicted Profit: ${prediction[0]:,.2f}"

# Define Gradio interface
inputs = [
    gr.Number(label="R&D Spend")
]

output = gr.Textbox(label="Prediction")

gr.Interface(fn=predict_profit, inputs=inputs, outputs=output, title="Startup Profit Predictor").launch()
```
