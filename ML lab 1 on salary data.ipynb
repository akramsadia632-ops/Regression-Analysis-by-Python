```python
from sklearn.model_selection import train_test_split
```

```python
# we want to predict 
# it should predict what will be the salary of person
# it i a problem of regression
# we can use linear regression model
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#we made scatter plot to check how much data is on X and on Y
```

```python
plt.scatter(df["YearsExperience"], df["Salary"], color='blue', marker='o')
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Experience vs Salary")
plt.show()
```

```python
# put input variables in X
# and output whch are independent variables in Y
```

```python
X=df[["YearsExperience"]]
Y=df["Salary"]
```

```python
X_tran, X_test, Y_tran,Y_test=train_test_split(X,Y, test_size=0.2,random_state=42)
```

```python
Lr=LinearRegression()
Lr.fit(X_tran,Y_tran)
```

```python
Y_pred=Lr.predict(X_test)
```

```python
#finding evaluation matrix
```

```python
r2_score=r2_score(Y_test,Y_pred)
MSE=mean_squared_error(Y_test,Y_pred)
```
